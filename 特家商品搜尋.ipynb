```python
import requests
import json
import re
from pprint import pprint
import pandas as pd
st = '2022-02-19'
et = '2022-05-20'
# keyword = "GreenPan 不沾鍋"
# keyword = "OHYAMA 碎紙機"
# keyword = "(JOYOUNG OR 九陽) 豆漿機 (開團 OR 團購 OR 訂購)"
# keyword = "Winix 清淨機"
# keyword = "Winix 除濕機"
# keyword = "IGRASS 大理石"
# keyword = "IRIS 織物清洗機"
# keyword = 'WEDGWOOD 寢具'
# keyword = "WMF 陶鋼鍋"
# keyword = "ricopa IH 電磁爐"
# keyword ="P F Candles CO 蠟燭"
# keyword = "小V 鬆餅機 (開團 OR 團購 OR 訂購)"
keyword = 'OHYAMA 碎紙機'
form = {
    'st': st,
    'et': et,
    'q': json.dumps([f"{keyword}"]),
    'vw': 'multiplechannel',
    "nation": 'TW',
    'channels': '["IG"]',
    'panels': '["LC","TC","HP","WC","HC","TL"]',
    'key': "1d7016605dcbd5fe717b61cd39c737bba4cc3095345557313eff8c92de45bd53"
}
r = requests.get('https://api.qsearch.cc/api/trend/v1/posts', params=form)
# pprint(r.json()['data'][0]['ig_raw'])
data = []
for i in r.json()['data'][0]['ig_raw']:
    page_name = i['from_name']
    page_url = 'https://www.instagram.com/'+i['username']
    fan_counts = i['follower_count']
    post_url = i['permalink']
    comments = i['coment_count']
    like = i['like_count']
    created_at = i['created_time']
    group_urls = ",".join(re.findall("(https?://[^\s]+)[^\S]",i['text']))
    data.append(
        {
            'page_name':page_name,
            'page_url':page_url,
            'fan_counts':fan_counts,
            'post_url':post_url,
            'comments':comments,
            'like':like,
            'created_at':created_at,
            'group_url':group_urls
        }
    )
pd.DataFrame(data).to_clipboard(sep=",",index=False)
pd.DataFrame(data)
```
